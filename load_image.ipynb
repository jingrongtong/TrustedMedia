# Load Image Header

```python
import exifread
import PIL.ExifTags
```

```python
img = PIL.Image.open(img_path)
img.verify()
```

```python
exif_data = img._getexif()
exif_data
```

```python
# Open image file for reading (binary mode)
f = open(img_path, 'rb')

# Return Exif tags
tags = exifread.process_file(f)
```
